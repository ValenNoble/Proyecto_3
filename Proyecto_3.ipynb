##¿Cuál es la diferencia en los niveles de pobreza entre Uruguay y Paraguay?"

##Introducción
El objetivo de este informe es analizar y comparar la diferencia en los niveles de pobreza entre Uruguay y Paraguay, durante el periodo de 2006 a 2019. Para realizar este objetivo, se empleará una visualización de datos que responde a la pregunta de investigación: **"¿Cuál es la diferencia en los niveles de pobreza entre Uruguay y Paraguay?"**

En este caso, elegí como país de comparación a Paraguay. Pienso que este es un país similar a Uruguay y que también esta emergiendo economicamentemente sin que se le de mucha atención.

La tasa de pobreza es un indicador utilizado en la evaluación de las condiciones económicas y sociales de una población. Representa el porcentaje de personas que viven por debajo del umbral de pobreza, es decir, aquellos cuyos ingresos no son suficientes para satisfacer sus necesidades básicas, como alimentación, vivienda, educación y atención médica. Este indicador ayuda a comprender la distribución de la riqueza y evaluar la efectividad de las políticas gubernamentales destinadas a reducir la pobreza.

La tasa de pobreza se calcula mediante la comparación de los ingresos de los individuos o familias con un umbral de pobreza predefinido. Si los ingresos caen por debajo de este umbral, las personas se consideran en situación de pobreza. Este cálculo puede variar según el país y la metodología utilizada, pero suele ser una herramienta utilizada para medir el bienestar económico y social de la población.

En este informe, se utilizarán datos de fuentes confiables para analizar y visualizar la evolución de la tasa de pobreza en Uruguay y Paraguay, lo que permitirá obtener información sobre la situación económica de ambos países durante el período especificado anteriormente.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de Uruguay y Paraguay
uruguay_data = pd.read_csv("/content/Proyecto_3/Uruguay_Datos.csv", encoding='latin-1')
paraguay_data = pd.read_csv("/content/Proyecto_3/UNdata_Export_20231102_180657312.csv", encoding='latin-1')

# Convertir la columna 'Year' en el DataFrame de Paraguay a tipo int
paraguay_data = paraguay_data[paraguay_data['Year'].str.isnumeric()]
paraguay_data['Year'] = paraguay_data['Year'].astype(int)

uruguay_dato_filtrado = uruguay_data[(uruguay_data['año']>=2006)&(uruguay_data['año']<=2019)]
paraguay_dato_filtrado = paraguay_data[(paraguay_data['Year']>=2006)&(paraguay_data['Year']<=2019)]
```

```python
# Crear una gráfica de línea que muestre la diferencia en los niveles de pobreza entre Uruguay y Paraguay
plt.figure(figsize=(10, 6))
plt.plot(uruguay_dato_filtrado['año'], uruguay_dato_filtrado['valor'], label='Uruguay', color='blue')
plt.plot(paraguay_dato_filtrado['Year'], paraguay_dato_filtrado['Value'], label='Paraguay', color='green')
plt.xlabel('Año')
plt.ylabel('Nivel de Pobreza')
plt.title('Diferencia en los Niveles de Pobreza entre Uruguay y Paraguay (2006-2019)')
plt.legend()
plt.grid(True)

plt.show()
```

Con el gráfico podemos visualizar la diferencia entre los niveles de pobreza entre Uruguay y Paraguay, lo cual aborda la pregunta de investigación planteada debido a que proporciona una representación clara y comparativa de la evolución de la tasa de pobreza en ambos países durante el período de 2006 a 2019. La gráfica de líneas muestra las tendencias en los niveles de pobreza a lo largo de los años, lo que permite identificar diferencias significativas en el progreso de ambos países en la reducción de la pobreza.

Este gráfico es veraz, ya que se basa en datos reales de dos fuentes confiables. Es funcional, debido a que proporciona una comparación efectiva y permite a los usuarios identificar rápidamente las diferencias entre Uruguay y Paraguay. Por otra parte, la visualización también es atractiva visualmente, haciendo que la información sea accesible y comprensible. La leyenda y etiquetas proporcionan contexto y claridad. En términos de perspicacia, la gráfica permite a los investigadores y tomadores de decisiones entender mejor la dinámica de la pobreza en estos dos países, lo que puede ser útil para la formulación de políticas y análisis comparativos.


##Conclusión
En el gráfico podemos visualizar los niveles de pobreza entre Uruguay y Paraguay durante el período de 2006 a 2019 proporciona una serie de observaciones significativas. Duarte estos años, se puede observar que Uruguay experimentó una disminución constante en la tasa de pobreza, pasando de 3.5% en 2006 a 0.5% en 2019, con fluctuaciones menores en años posteriores. En contraste, con Paraguay se ve una disminución de la pobreza, pero en un ritmo más lento, pasando de 2.4% en 2006 a 0.2% en 2019.

Con esto podemos ver que Uruguay ha tenido un éxito notable en la reducción de la pobreza durante este período, logrando tasas de pobreza consistentemente más bajas que Paraguay. Si bien ambos países han experimentado mejoras, Uruguay ha mantenido una tendencia más sólida hacia la disminución de la pobreza.
