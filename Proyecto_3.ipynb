¿Cuál es la diferencia en los niveles de pobreza entre Uruguay y Paraguay?"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de Uruguay y Paraguay
uruguay_data = pd.read_csv("/content/Proyecto_3/Uruguay_Datos.csv", encoding='latin-1')
paraguay_data = pd.read_csv("/content/Proyecto_3/UNdata_Export_20231102_180657312.csv", encoding='latin-1')

# Convertir la columna 'Year' en el DataFrame de Paraguay a tipo int
paraguay_data = paraguay_data[paraguay_data['Year'].str.isnumeric()]
paraguay_data['Year'] = paraguay_data['Year'].astype(int)

# Fusionar los datos en función del año
merged_data = uruguay_data.merge(paraguay_data, left_on='año', right_on='Year')

# Renombrar las columnas
merged_data = merged_data.rename(columns={'valor_x': 'Pobreza Uruguay', 'Value': 'Pobreza Paraguay'})
```

```python
# Fusión de datos en función del año
merged_data = uruguay_data.merge(paraguay_data, left_on='año', right_on='Year')

# Renombrar las columnas en consecuencia
merged_data = merged_data.rename(columns={'valor': 'Pobreza Uruguay', 'Value': 'Pobreza Paraguay'})
```

```python
# Crear una gráfica de líneas
plt.figure(figsize=(10, 6))
plt.plot(merged_data['año'], merged_data['Pobreza Uruguay'], label='Uruguay')
plt.plot(merged_data['año'], merged_data['Pobreza Paraguay'], label='Paraguay')

# Etiquetas, título y leyenda
plt.xlabel('Año')
plt.ylabel('Tasa de Pobreza')
plt.title('Diferencia en los Niveles de Pobreza entre Uruguay y Paraguay (2006-2019)')
plt.legend()

plt.show()
```
